```python
import matplotlib.pyplot as plt

def configurar_grafica():
    plt.xlim(-0.2, 1.2)
    plt.ylim(-0.2, 1.2)
    plt.xticks([0, 1])
    plt.yticks([0, 1])
    plt.grid(True)

def graficar_puntos(x,y):    
    for xi, yi in zip(x, y):
        color = 'green' if yi == 1 else 'red'
        plt.scatter(xi[0], xi[1], c=color, s=200)

def graficar_linea(w0,w1,b):
    x_values = []
    y_values = []
    for i in range(140):
        x = -0.2 + i / 100 
        y = -(w0 * x + b) / w1 if w1 != 0 else 0
        x_values.append(x)
        y_values.append(y)
    
    plt.fill_between(x_values, -1, y_values,  color='red',   alpha=0.2)
    plt.fill_between(x_values, y_values, 1.2, color='green', alpha=0.2)
    plt.show()

x = [[0,0], [0,1], [1,0], [1,1]]
y = [0, 1, 1, 1]

w0 = 0
w1 = 0
b  = 0

def f(x):
    linear_combination = w0 * x[0] + w1 * x[1] + b
    return 1 if linear_combination >= 0 else 0

configurar_grafica()
graficar_puntos(x, y)
graficar_linea(0.2, .1, -.05)
```

```python

import csv
import random

def generar_datos_credito(ruta_csv, buenos=50, malos=50):
    with open(ruta_csv, mode='w', newline='', encoding='utf-8') as f:
        escritor = csv.writer(f)
        escritor.writerow(["sueldo", "antiguedad", "cumplio"])
        
        # Buenos clientes
        for _ in range(buenos):
            sueldo = random.randint(500000, 1000000)  # sueldos altos
            antiguedad = random.randint(24, 120)       # antigüedad larga
            cumplio = 1
            escritor.writerow([sueldo, antiguedad, cumplio])

        # Malos clientes
        for _ in range(malos):
            sueldo = random.randint(100000, 400000)   # sueldos bajos
            antiguedad = random.randint(1, 12)         # antigüedad corta
            escritor.writerow([sueldo, antiguedad, cumplio])

generar_datos_credito("50.3.creditos.csv", buenos=100, malos=100)
```
