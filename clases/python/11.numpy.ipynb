```python
import numpy as np
import matplotlib.pyplot as plt

# Crear datos
x = np.linspace(-np.pi, np.pi, 200)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Graficar
plt.figure(figsize=(8,5))
plt.plot(x, y_sin, label='sin(x)')
plt.plot(x, y_cos, label='cos(x)')
plt.title("Funciones Seno y Coseno entre -π y π")
plt.xlabel("x (radianes)")
plt.ylabel("Valor")
plt.legend()
plt.grid(True)
# plt.show()
y_cos_shifted = np.cos(x + np.pi/4)
plt.plot(x, y_cos_shifted, label='cos(x + π/4)')
plt.legend()
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 100)
y = np.linspace(-np.pi, np.pi, 100)

X, Y = np.meshgrid(x, y)
Z = np.sin(X) + np.cos(Y)

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title("Superficie: z = sin(x) + cos(y)")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo'],
    'Ventas': [3050, 3560, 3900, 4250, 4800],
    'Gastos': [2000, 2100, 2500, 2600, 3000]
}

df = pd.DataFrame(data)
print(df)
```

```python
plt.plot(df['Mes'], df['Ventas'], label='Ventas', marker='o')
plt.plot(df['Mes'], df['Gastos'], label='Gastos', marker='s')
plt.title("Ventas vs Gastos por mes")
plt.xlabel("Mes")
plt.ylabel("Monto ($)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
df.plot(x='Mes', y=['Ventas', 'Gastos'], kind='bar')
plt.title("Comparación mensual")
plt.ylabel("Monto ($)")
plt.show()
```

```python
plt.pie(df['Ventas'], labels=df['Mes'], autopct='%1.0f%%', startangle=90)
plt.title("Porcentaje de ventas por mes")
plt.show()
```

```python
df['Beneficio'] = df['Ventas'] - df['Gastos']

plt.bar(df['Mes'], df['Beneficio'], color='green')
plt.title("Beneficio mensual")
plt.xlabel("Mes")
plt.ylabel("Beneficio ($)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')

# make a stream function:
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
# make U and V out of the streamfunction:
V = np.diff(Z[1:, :], axis=1)
U = -np.diff(Z[:, 1:], axis=0)

# plot:
fig, ax = plt.subplots()

ax.streamplot(X[1:, 1:], Y[1:, 1:], U, V)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# Make data for a double helix
n = 50
theta = np.linspace(0, 2*np.pi, n)
x1 = np.cos(theta)
y1 = np.sin(theta)
z1 = np.linspace(0, 1, n)
x2 = np.cos(theta + np.pi)
y2 = np.sin(theta + np.pi)
z2 = z1

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.fill_between(x1, y1, z1, x2, y2, z2, alpha=0.5)
ax.plot(x1, y1, z1, linewidth=2, color='C0')
ax.plot(x2, y2, z2, linewidth=2, color='C0')

ax.set(xticklabels=[], yticklabels=[], zticklabels=[])


plt.show()
```

```python
import numpy as np
import plotly.graph_objects as go

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y)])
fig.update_layout(title="Superficie 3D - función seno", scene=dict(
    xaxis_title='Eje X',
    yaxis_title='Eje Y',
    zaxis_title='Z = sin(sqrt(X² + Y²))'
))
fig.show()
```
